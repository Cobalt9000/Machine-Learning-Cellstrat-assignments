Q4 - What is the purpose of StandardScaler class?

Answer: The main idea is to normalize/standardize i.e. μ = 0 and σ = 1 your features/variables/columns of X , individually, before applying any machine learning model. StandardScaler() will normalize the features i.e. each column of X, INDIVIDUALLY, so that each column/feature/variable will have μ = 0 and σ = 1.

StandardScaler removes the mean and scales each feature/variable to unit variance. This operation is performed feature-wise in an independent way. StandardScaler can be influenced by outliers (if they exist in the dataset) since it involves the estimation of the empirical mean and standard deviation of each feature.

for degree 2:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
insuranceData = pd.read_csv ('Insurance.csv')

X = insuranceData.iloc [:,0:1].values
y = insuranceData.iloc [:, 1].values

from sklearn.linear_model import LinearRegression
linregressor = LinearRegression()
linregressor.fit(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures # Features

# First create a polynomial
polyfeature = PolynomialFeatures(degree = 2)
Poly_Matrix = polyfeature.fit_transform(X)

# Now fit the polynomial matrix into linear regressor
linregressor2 = LinearRegression()
linregressor2.fit(Poly_Matrix, y)
```

```python
# Visualising the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, linregressor.predict(X), color = 'blue')
plt.title('Insurance Premium - Linear')
plt.xlabel('Age')
plt.ylabel('Premium')
plt.show()

# Visualising the Polynomial Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, linregressor2.predict(polyfeature.fit_transform(X)), color = 'green')
plt.title('Insurance Premium - Polynomial')
plt.xlabel('Age')
plt.ylabel('Premium')
plt.show()
```

Q3 - For degree 3 polynomial regression, what is the predicted insurance premium for ages 29, 45, 53 and 80 ?

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

insuranceData = pd.read_csv ('Insurance.csv')

X = insuranceData.iloc [:,0:1].values
y = insuranceData.iloc [:, 1].values

from sklearn.linear_model import LinearRegression
linregressor = LinearRegression()
linregressor.fit(X, y)

from sklearn.preprocessing import PolynomialFeatures # Features

# First create a polynomial
polyfeature = PolynomialFeatures(degree = 3)
Poly_Matrix = polyfeature.fit_transform(X)

# Now fit the polynomial matrix into linear regressor
linregressor2 = LinearRegression()
linregressor2.fit(Poly_Matrix, y)

# Visualising the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, linregressor.predict(X), color = 'blue')
plt.title('Insurance Premium - Linear')
plt.xlabel('Age')
plt.ylabel('Premium')
plt.show()

# Visualising the Polynomial Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, linregressor2.predict(polyfeature.fit_transform(X)), color = 'green')
plt.title('Insurance Premium - Polynomial')
plt.xlabel('Age')
plt.ylabel('Premium')
plt.show()
```

The predicted insurance premium for ages 29, 45, 53 and 80 are:

```python
val1 = np.array([29])
val_new1=val1.reshape(-1,1)
predictionlinear1 = linregressor.predict(val_new1)
predictionPoly1 = linregressor2.predict (polyfeature.fit_transform(val_new1))

print ("linear prediction of age 29 = ", predictionlinear1)
print ("polynomial prediction of age 29 = ", predictionPoly1)

val2 = np.array([45])
val_new2=val2.reshape(-1,1)
predictionlinear2 = linregressor.predict(val_new2)
predictionPoly2 = linregressor2.predict (polyfeature.fit_transform(val_new2))

print ("linear prediction of age 45 = ", predictionlinear2)
print ("polynomial prediction of age 45 = ", predictionPoly2)


val3 = np.array([53])
val_new3=val3.reshape(-1,1)
predictionlinear3 = linregressor.predict(val_new3)
predictionPoly3 = linregressor2.predict (polyfeature.fit_transform(val_new3))

print ("linear prediction of age 53 = ", predictionlinear3)
print ("polynomial prediction of age 53 = ", predictionPoly3)


val4 = np.array([80])
val_new4=val4.reshape(-1,1)
predictionlinear4 = linregressor.predict(val_new4)
predictionPoly4 = linregressor2.predict (polyfeature.fit_transform(val_new4))

print ("linear prediction of age 80 = ", predictionlinear4)
print ("polynomial prediction of age 80 = ", predictionPoly4)
```
